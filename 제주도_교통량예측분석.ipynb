```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/데이터마이닝/train.csv")
test = pd.read_csv("/content/drive/MyDrive/데이터마이닝/test.csv")
```

```python
df.info() # object 열 확인 -> 범주형데이터로 수정필요 (+ base_date)
```

```python
df = df.drop(['id'], axis=1) # 'id'열 삭제
```

```python
dummies = pd.get_dummies(df.start_turn_restricted) # 원핫인코딩 처리
merged = pd.concat([df,dummies], axis='columns')
df = merged
df = df.drop(['start_turn_restricted'], axis=1)
df.rename(columns={'없음':'start_turn_restricted_N','있음':'start_turn_restricted_Y'}, inplace=True)
```

```python
dummies = pd.get_dummies(df.end_turn_restricted) # 원핫인코딩 처리
merged = pd.concat([df,dummies], axis='columns')
df = merged
df = df.drop(['end_turn_restricted'], axis=1)
df.rename(columns={'없음':'end_turn_restricted_N','있음':'end_turn_restricted_Y'}, inplace=True)
```

```python
df['base_date'] = pd.to_datetime(df['base_date'],format='%Y%m%d') # 날짜 타입 변경
```

```python
from sklearn.preprocessing import LabelEncoder # 라벨인코딩

le = LabelEncoder()
le_encoded = le.fit_transform(df.base_date)
new_cat = pd.DataFrame(le_encoded, columns = ['base_date'])
df = df.drop(['base_date'], axis=1)
merged = pd.concat([df,new_cat], axis='columns')
df = merged
```

```python
le = LabelEncoder() # 라벨인코딩
le_encoded = le.fit_transform(df.day_of_week)
new_cat = pd.DataFrame(le_encoded, columns = ['day_of_week'])
df = df.drop(['day_of_week'], axis=1)
merged = pd.concat([df,new_cat], axis='columns')
df = merged
```

```python
le = LabelEncoder() # 라벨인코딩
le_encoded = le.fit_transform(df.road_name)
new_cat = pd.DataFrame(le_encoded, columns = ['road_name'])
df = df.drop(['road_name'], axis=1)
merged = pd.concat([df,new_cat], axis='columns')
df = merged
df.info()
```

```python
le = LabelEncoder() # 라벨인코딩
le_encoded = le.fit_transform(df.start_node_name)
new_cat = pd.DataFrame(le_encoded, columns = ['start_node_name'])
df = df.drop(['start_node_name'], axis=1)
merged = pd.concat([df,new_cat], axis='columns')
df = merged
df.info()
```

```python
le = LabelEncoder() # 라벨인코딩
le_encoded = le.fit_transform(df.end_node_name)
new_cat = pd.DataFrame(le_encoded, columns = ['end_node_name'])
df = df.drop(['end_node_name'], axis=1)
merged = pd.concat([df,new_cat], axis='columns')
df = merged
df.info()
```

```python
df['target'] = pd.cut(df.target, bins=[0,41.1,113], labels=[1,0]) # 종속변수 0,1로 바꿈
```

```python
x = df.drop('target', axis=1) # x:독립변수, y:종속변수
y = df['target']
```

```python
from sklearn.preprocessing import MinMaxScaler # 독립변수 min-max scaling

scaler = MinMaxScaler()
x_scaler = scaler.fit_transform(x)
x_scaler = pd.DataFrame(x_scaler, columns = x.columns)
x_scaler
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size = 0.2, random_state = 0)
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=5, random_state=0)

dt.fit(x_train, y_train)

pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

print("Accuracy: ", accuracy_score(y_test,pred))
print("Precision: ", precision_score(y_test,pred))
print("Recall: ", recall_score(y_test,pred))
print("F1: ", f1_score(y_test,pred))
```

```python
# feature importance 확인
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(dt, random_state=0).fit(x_test, y_test)
eli5.show_weights(perm, feature_names = x_test.columns.tolist())
```
